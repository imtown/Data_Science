## 1. Import

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

## 2. Use Style

```python
plt.style.available
plt.style.use('ggplot')
plt.plot();
```

```python
x = [1,2,3,4]
y = [10, 8, 6, 9]
plt.plot(x,y, color = 'red');
```

## 3. Pyplot API

```python
x = np.linspace(0,10, 1000)
# line sin
plt.plot(x, np.sin(x), color = 'Blue', linestyle = 'dashed', 
         label = 'Sin(X)');
# line cos
plt.plot(x, np.cos(x), color = 'red', label = 'Cos(x)')
plt.title('Sin(X) and Cos(X)')
plt.xlabel('Bien X')
plt.ylabel('Sin(X)')
#plt.xlim(0,10)
#plt.ylim(-0.75, 1.5)
plt.axis([1,10, -4, 2])
plt.legend();
```

## 4. Object-Oriented API

```python
x = [1,2,3,4]
y = [11,22,33,44]
fig, ax = plt.subplots(figsize = (5,5)) # w and h

ax.plot(x,y);

ax.set(title = 'A simple plot',
      xlabel = 'X-axis',
      ylabel = 'Y-axis');
```

## 6. Scatter

```python
# Simple Scatter Chart
plt.scatter(x, np.exp(x), color='steelblue', marker='o', s=50);
```

```python
# OO API Scatter Chart
rng = np.random.RandomState(0)

x = rng.randn(100)
y = rng.randn(100)
colors = rng.rand(100)
sizes = 1000*rng.rand(100)

fig, ax = plt.subplots()
img1 = ax.scatter(x,y,s = sizes, c = colors, cmap = 'viridis', 
           alpha = 0.3);

fig.colorbar(img1);
```

## 7. Bar chart

```python
sales = ['Branch', 'Customer', 'Item', 'Qty']
val1 = [15, 41, 28, 34]
val2 = [18, 30, 30, 10]
```

```python
#simple bar
plt.bar(sales, val1, align='edge',color='royalblue', width=0.8, edgecolor='b', linewidth=1.3, alpha=0.8, 
       linestyle='--', label='Sales Registe', visible=True);
plt.legend();
```

```python
# Simple horizontal bar
plt.figure(figsize=(10,5))
sales_len = np.arange(len(sales))
width = 0.3
plt.barh(sales_len, val1, width, color='cornflowerblue', linewidth=1.5, label='Value_1mth');
plt.barh(sales_len + width, val2, width, color='lightskyblue', linewidth=1, label='Value_2mth');
plt.legend();
#plt.xticks(sales_len + width, sales);
```

```python
# Vertical bar chart API
soft_drink_prices = {'Coca' : 10,
                    'Pessi' : 12,
                    'Sprite' : 8}
fig, ax = plt.subplots()
ax.bar(soft_drink_prices.keys(), soft_drink_prices.values())
ax.set(title = 'Bach Hoa Xanh\'s Soft Drink Prices',
      ylabel = 'Price ($)');
```

```python
# Vertical bar chart API
fid, ax = plt.subplots()
ax.barh(list(soft_drink_prices.keys()),
        list(soft_drink_prices.values()));
```

## 8. Histogram

```python
np.random.seed(42)
student_hight = np.random.normal(170, 15, size=250)
student_hight[:10]
```

```python
plt.hist(student_hight, bins = 30);
```

```python
bins = [140, 150, 160, 170, 175, 180]
plt.hist(student_hight, bins,rwidth=0.8, histtype='bar', 
         orientation='horizontal', color='c');

plt.title('Student_Height');
plt.ylabel('Centimeter');
```

## 9. Subplots

```python
fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(nrows = 2,
                      ncols = 2,
                      figsize = (10, 5))
ax1.plot(x, x/2);
ax1.plot(x, x*2)

ax2.scatter(np.random.random(10), np.random.random(10));

ax3.bar(soft_drink_prices.keys(), soft_drink_prices.values());

ax4.hist(student_hight, bins = 30);
```

```python
fig, ax = plt.subplots(nrows = 2,
                      ncols = 2,
                      figsize = (10, 5))
ax[0,0].plot(x, x/2);
ax[0,0].plot(x, x*2)

ax[0,1].scatter(np.random.random(10), np.random.random(10));

ax[1,0].bar(soft_drink_prices.keys(), soft_drink_prices.values());

ax[1,1].hist(student_hight, bins = 30);
```

## 10. Pie Chart

```python
product = ['image', 'feature', 'mix', 'use', 'service']
income = [100, 135, 200, 98, 170]
```

```python
plt.figure(figsize=(5,5))
explode = [0,0,0.1,0,0]
color = ['b', 'r', 'seagreen', 'c', 'lime']
plt.pie(income, labels=product, explode=explode, colors=color, 
        autopct='%.2f%%', shadow=True, textprops={'fontsize':15},startangle=360);
plt.legend(loc=1);
```

# Project: Data Visualization - Califotnia Cities 

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
cities = pd.read_csv('./DATA INPUT/california_cities.csv')
cities.head()
```

```python
lat, lon = cities['latd'], cities['longd']

population, area = cities['population_total'], cities['area_total_km2']

plt.figure(figsize = (8,6))

# Plots
plt.scatter(lat, lon,
            c = np.log10(population),
            cmap = 'viridis',
           s = area,
           linewidths = 0,
           alpha = 0.5);
plt.axis('equal');
plt.xlabel('Latitude')
plt.ylabel('Longlatitude')
plt.colorbar(label = 'log${10}$(pupolation)');

plt.title('Califotnia Cities: Population and Area')

# savefig() to backup chart
plt.savefig('Califotnia_Cities_Project')
```
