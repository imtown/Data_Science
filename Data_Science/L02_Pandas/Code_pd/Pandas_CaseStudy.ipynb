# Datetime and Time Series

```python
import pandas as pd
df = pd.read_csv('./DATA INPUT/Electric_Production.csv', parse_dates=['DATE'])
```

```python
df.head()
```

```python
df.loc[0, 'DATE']
```

### Link [Datetime fortmatting codes] 
https://docs.python.org/3/library/datetime.html#strftime-and-strptime-behavior

```python
df['DATE'] = pd.to_datetime(df['DATE'], format='%Y-%m-%d  %I-%p')
```

```python
df['DATE'] 
```

```python
df.loc[0, 'DATE'].day_name()
```

```python
df['DATE'].dt.day_name()
```

```python
df['DayOfWeek'] = df['DATE'].dt.day_name()
```

```python
df
```

```python
filt = (df['DATE'] >= pd.to_datetime('2015-05-01')) & (df['DATE'] < pd.to_datetime('2017-06-01'))
df.loc[filt]
```

```python
df.set_index('DATE', inplace=True)
```

```python
df.head()
```

```python
df['2017' : '2018']
```

```python
df['2017' : '2018']['IPG2211A2N']
```

```python
df['2017' : '2018'].min()
```

```python
df['2017' : '2018'].max()
```

```python
df['2017' : '2018'].mean()
```

### Link [DateOffset objects]
https://pandas.pydata.org/pandas-docs/stable/user_guide/timeseries.html#dateoffset-objects

```python
hight = df['2015':'2017']['IPG2211A2N'].resample('Q').max()
hight.head(10)
```
