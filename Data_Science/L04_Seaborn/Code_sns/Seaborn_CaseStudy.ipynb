## Correlation Scatter Plot Project with Nutrition

```python
# Load Data
df_nut = pd.read_csv('./DATA INPUT/cereal.csv')
df_nut.head()
```

```python
df_nut.info()
```

```python
# Drop some columns
df_drop = ['shelf', 'weight', 'cups', 'rating']
df_nutri = df_nut.drop(df_drop, axis=1)
df_nutri.head()
```

```python
# use corr() to correlation data
df_corr = df_nutri.corr()
df_corr
```

```python
# Return bool type
df_nones = np.ones_like(df_corr, dtype=bool)
df_nones
```

```python
# Create mask to drop same values
mask = np.triu(df_nones)
mask
```

```python
# Heatmap
sns.heatmap(data=df_corr, mask=mask, cmap='coolwarm');
```

```python
# Remove middile line of mask
add_middle1 = mask[1:, :-1]
add_middle1
```

```python
# Remove middile line of correlation
add_middle2 = df_corr.iloc[1: , :-1]
add_middle2
```

```python
fig, ax = plt.subplots(figsize=(10,7))
sns.heatmap(data=add_middle2, mask=add_middle1, annot=True, 
            fmt='.2f',linewidths=.5, linecolor='w', vmin=1, vmax=-1);

ax.set_title('Nutrition Correlation', loc='left', fontsize=20, color='m');20
```
